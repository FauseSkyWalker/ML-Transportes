```python

# Carregar os arquivos
df1 = pd.read_csv('/home/fause/ML-Transportes/DadosPRF/acidentes2025_todas_causas_tipos.csv', encoding='iso-8859-1', sep=';')
df2 = pd.read_csv('/home/fause/ML-Transportes/DadosPRF/acidentes2025PorPessoa.csv', encoding='iso-8859-1', sep=';')
df3 = pd.read_csv('/home/fause/ML-Transportes/DadosPRF/datatran2025PorOcorrencia.csv', encoding='iso-8859-1', sep=';')
```

```python
# Obter conjuntos de colunas
colunas1 = set(df1.columns)
colunas2 = set(df2.columns)
colunas3 = set(df3.columns)
```

```python
# Mostrar colunas exclusivas de cada uma
print("üî∏ Colunas exclusivas do df1:")
print(colunas1 - colunas2 - colunas3)
```

```python
print("\nüî∏ Colunas exclusivas do df2:")
print(colunas2 - colunas1 - colunas3)
```

```python
print("\nüî∏ Colunas exclusivas do df3:")
print(colunas3 - colunas1 - colunas2)
```

```python
# Mostrar colunas em comum entre as tr√™s
comuns_todas = colunas1 & colunas2 & colunas3
print("\n‚úÖ Colunas em comum nas 3 planilhas:")
print(comuns_todas)
```

```python
# Mostrar colunas em comum entre pelo menos duas
comuns_12 = colunas1 & colunas2
comuns_13 = colunas1 & colunas3
comuns_23 = colunas2 & colunas3

comuns_duas = (comuns_12 | comuns_13 | comuns_23) - comuns_todas
print("\nüîÅ Colunas em comum entre apenas 2 planilhas:")
print(comuns_duas)
```
