```python
import pandas as pd

# 1. Isolar a Marca Principal (a parte antes da primeira '/')
# Usamos .str.split('/', n=1) para dividir apenas na primeira barra,
# e .str[0] para pegar o primeiro elemento (a marca).
df['Marca_Principal'] = df['marca'].str.split('/', n=1).str[0]

# 2. Agrupar pela Marca Principal e listar todas as variações de nomes completos
# unique() retorna um array com as variações.
variacoes_por_marca = df.groupby('Marca_Principal')['marca'].unique()

print("--- Variações de nomes de veículos por Marca Principal para Inspeção ---")
print("Procure por nomes semelhantes que deveriam ser um só (e.g., VW/GOL e VW/GOL/GOL)\n")

# 3. Exibir apenas as marcas que possuem MÚLTIPLAS variações de nomes
# Usamos um loop para formatar a saída de forma legível
for marca, lista_variacoes in variacoes_por_marca.items():
    # Exibe apenas se houver mais de uma variação ou se for o caso de "Não Informado"
    if len(lista_variacoes) > 1 and marca != 'Não Informado':
        print(f"\n--- {marca} ({len(lista_variacoes)} variações) ---")
        # Itera sobre a lista de variações para exibir
        for variacao in lista_variacoes:
            print(f"- {variacao}")

# Se o resultado for muito grande, você pode limitar a visualização para as primeiras 100 marcas, por exemplo:
# for marca, lista_variacoes in variacoes_por_marca.head(100).items():
#     ... (o mesmo código acima)
```

```python
import pandas as pd
import re

# Função auxiliar para extrair o Nome do Modelo (e.g., FIESTA, KA)
def extrair_nome_modelo(modelo_grupo):
    if not modelo_grupo:
        return 'Não Informado'
    
    # 1. Pega a parte principal do modelo (antes da sub-divisão como /GOL 1.0)
    primeira_parte = modelo_grupo.split('/')[0].strip()
    
    if not primeira_parte:
        return 'Outros'

    # 2. Pega a primeira palavra, que é o nome do modelo (FIESTA, KA, GOL, etc.)
    nome_modelo = primeira_parte.split(' ')[0]
    
    # 3. Limpa e padroniza para maiúsculas
    nome_modelo = re.sub(r'[^A-Z0-9]', '', nome_modelo.upper())
    
    return nome_modelo if nome_modelo else 'Outros'

# 1. EXTRAÇÃO DAS CAMADAS HIERÁRQUICAS
df['Marca_Principal'] = df['marca'].str.split('/', n=1).str[0].str.strip()
df['Modelo_Grupo'] = df['marca'].str.split('/', n=1).str[1].fillna('').str.strip()
df['Nome_Modelo'] = df['Modelo_Grupo'].apply(extrair_nome_modelo)

# 2. AGRUPAMENTO E CONTAGEM
# Agrupa por Marca e Modelo e lista todas as variações detalhadas
df_agrupado = df.groupby(['Marca_Principal', 'Nome_Modelo'])['marca'].unique().apply(list).reset_index()
df_agrupado.columns = ['Marca_Principal', 'Nome_Modelo', 'Variações_Detalhes']

# Calcula o total de variações para a Marca Principal (para o cabeçalho)
total_variacoes_por_marca = df.groupby('Marca_Principal')['marca'].nunique()

# 3. IMPRESSÃO FORMATADA
print("Resultado da Hierarquia de Marcas e Modelos:\n")

for marca_principal in df_agrupado['Marca_Principal'].unique():
    
    # Cabeçalho da Marca
    total_variacoes = total_variacoes_por_marca.get(marca_principal, 0)
    print(f"--- {marca_principal} ({total_variacoes} variações) ---")
    
    # Filtrar os modelos para esta Marca
    modelos_da_marca = df_agrupado[df_agrupado['Marca_Principal'] == marca_principal].sort_values(by='Nome_Modelo')
    
    # Para cada Modelo dentro da Marca
    for index, row in modelos_da_marca.iterrows():
        nome_modelo = row['Nome_Modelo']
        variacoes_detalhes = row['Variações_Detalhes']
        count_modelo = len(variacoes_detalhes)
        
        # Sub-cabeçalho do Modelo
        print(f"               {nome_modelo} ({count_modelo} variações)")
        
        # Lista de Variações Detalhadas
        for variacao in variacoes_detalhes:
            print(f"- {variacao}")

    print("\n")
```

```python
import pandas as pd
import re

# Definição da função de extração da subárvore (Nome do Modelo)
def extrair_nome_modelo(modelo_grupo):
    """
    Extrai o primeiro nome do modelo de uma string de variação.
    Ex: 'FIESTA FLEX/FIESTA FLEX' -> 'FIESTA'
    """
    if not modelo_grupo:
        return 'NÃO INFORMADO'
    
    # 1. Pega a parte principal do modelo (antes da sub-divisão como /GOL 1.0)
    primeira_parte = modelo_grupo.split('/')[0].strip()
    
    if not primeira_parte:
        return 'OUTROS'

    # 2. Pega a primeira palavra, que é o nome do modelo (FIESTA, KA, GOL, etc.)
    nome_modelo = primeira_parte.split(' ')[0]
    
    # 3. Limpa e padroniza para maiúsculas
    nome_modelo = re.sub(r'[^A-Z0-9]', '', nome_modelo.upper())
    
    return nome_modelo if nome_modelo else 'OUTROS'

# --- APLICAÇÃO NO DATAFRAME (df) ---

# 1. Criação da coluna Marca_Arvore (Marca Principal)
# Extrai o que está antes da primeira barra.
df['Marca_Arvore'] = df['marca'].str.split('/', n=1).str[0].str.strip()

# 2. Criação da coluna Marca_Subarvore (Modelo Principal)
# a) Extrai temporariamente o restante da string para passar à função
modelo_grupo_temp = df['marca'].str.split('/', n=1).str[1].fillna('').str.strip()

# b) Aplica a função para obter o nome do modelo
df['Marca_Subarvore'] = modelo_grupo_temp.apply(extrair_nome_modelo)

# (Opcional) Mostra o resultado para confirmação
print(df[['marca', 'Marca_Arvore', 'Marca_Subarvore']].head(10))

# Salvar o DataFrame 'df' (que agora contém as colunas Marca_Arvore e Marca_Subarvore)
# O parâmetro index=False é crucial para não salvar a coluna de índice numérico do Pandas no CSV.

nome_arquivo = 'acidentes_pbic_2020_2025_condutor_marca.csv'
df.to_csv(nome_arquivo, index=False)

print(f"O novo arquivo CSV foi salvo com sucesso como '{nome_arquivo}' no diretório de trabalho atual.")
```
